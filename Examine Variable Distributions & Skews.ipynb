```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Read in Data
TicketData = pd.read_csv('Data/ProcessedTicketData.csv')
```

```python
TicketData['FV_delta'].plot.box()
```

```python
TicketData['FV_delta'].describe()
```

Variables we want to look at: 
'face_value', 'sold_out', 'days_to_show', 'num_blogs', 'num_news', 'num_reviews', 'discovery', 'familiarity', 'hotttnesss', 'num_years_active'

### Face_value

```python
TicketData['face_value'].plot.box()
print TicketData['face_value'].describe()
print("\nNumber of zero values: %s" % TicketData[TicketData['face_value'] <= 0]["face_value"].count())
```

```python
TicketData['face_value_log'] = np.log(TicketData['face_value'])
TicketData['face_value_log'].replace(to_replace=-float('Inf'), value=0, inplace=True)
TicketData['face_value_log'].plot.box()
TicketData['face_value_log'].describe()
```

### Days_to_show

```python
TicketData['days_to_show'].plot.box()
print TicketData['days_to_show'].describe()
print("\nNumber of zero values: %s" % TicketData[TicketData['days_to_show'] <= 0]["days_to_show"].count())
```

```python
# Take the log of the days_to_show values
TicketData['days_to_show_log'] = np.log(TicketData['days_to_show'])
# Convert negative infinity values to zero
TicketData['days_to_show_log'].replace(to_replace=-float('Inf'), value=0, inplace=True)
TicketData['days_to_show_log'].plot.box()
TicketData['days_to_show_log'].describe()
```
